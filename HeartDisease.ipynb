```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns



cleveland = pd.read_csv('Heart_Disease_Prediction.csv')

print( 'Shape of DataFrame:', cleveland.shape)
#cleveland.describe()

cleveland.columns = ['Age','Sex','Chest_pain_type', 'Blood_pressure','Cholesterol','Fasting_blood_sugar_over_120','EKG_results',
'Max_heart_rate_achieved','Exercise_angina','ST_depression','Slope_of_ST','Number_of_vessels_fluro',
'Thallium','Heart_disease']
cleveland['Sex']=cleveland['Sex'].replace(to_replace=[0,1],
           value=['femal','male'])

cleveland['Chest_pain_type']=cleveland['Chest_pain_type'].replace(to_replace=[1,2,3,4],
           value=['TA','ATA','NAP','ASY'])

cleveland['Heart_disease']=cleveland['Heart_disease'].replace(to_replace=['Absence','Presence'],
           value=['No heart Disease','Heart Disease'])

#plotting histogram of all numeric values
cleveland.hist(figsize = (12, 12))
plt.show()
```

```python
#How many people have heart disease and how many people doesn't have heart disease?

plt.pie(cleveland.Heart_disease.value_counts(), autopct='%1.2f%%',  startangle=180, radius=1.5)
plt.legend(["Disease", "No disease"]);
```

```python
#People of which sex has most heart disease?
print(pd.crosstab(cleveland.Heart_disease,cleveland.Sex))
pd.crosstab(cleveland.Heart_disease,cleveland.Sex).plot(kind="bar",figsize=(20,6))
plt.title("Heart Disease Frequency for Sex")
```

```python
#Number of male is more than double in our dataset than female.

#More than 55% male has heart disease and 25% female has heart disease.
```

```python
#Average Age
print(pd.crosstab(cleveland.Heart_disease,cleveland.Sex,values=cleveland.Age, aggfunc=np.average))
pd.crosstab(cleveland.Heart_disease,cleveland.Sex,values=cleveland.Age, aggfunc=np.average).plot(kind="bar",figsize=(20,6))
```

```python
#People of which sex has which type of chest pain most?
print(pd.crosstab(cleveland.Sex, cleveland.Chest_pain_type))
pd.crosstab(cleveland.Sex, cleveland.Chest_pain_type).plot(kind = 'bar', color = ['coral', 'lightskyblue', 'plum', 'khaki'])
plt.title('Type of chest pain for sex')
plt.xlabel('Sex');
```

```python
#Most of male has type ASY chest pain and least of Male has type TA pain.

#in case of Female type ASY and type NAP percentage is almost same.
```

```python
#Age vs Maximum heart rate for people have heart disease.
#Age vs Maximum heart rate for people doesn't have heart disease.

plt.figure(figsize=(10,6))
plt.scatter(cleveland.Age[cleveland.Heart_disease=='Heart Disease'],  cleveland.Max_heart_rate_achieved[cleveland.Heart_disease=='Heart Disease'],  c="coral") 
plt.scatter(cleveland.Age[cleveland.Heart_disease=='No heart Disease'],  cleveland.Max_heart_rate_achieved[cleveland.Heart_disease=='No heart Disease'],  c="lightskyblue") 

plt.title("Heart Disease w.r.t Age and Max Heart Rate")
plt.xlabel("Age")
plt.legend(["Heart Disease", "No heart Disease"])
plt.ylabel("Max Heart Rate");
```

```python
#Heart Disease is heartbeat too fast or too slow
```

```python
#Age vs Serum Cholesterol for people have heart disease.
#Age vs Serum Cholesterol for people doesn't have heart disease.

plt.figure(figsize=(10,6))
plt.scatter(cleveland.Age[cleveland.Heart_disease=='Heart Disease'],  cleveland.Cholesterol[cleveland.Heart_disease=='Heart Disease'],  c="coral") 
plt.scatter(cleveland.Age[cleveland.Heart_disease=='No heart Disease'],  cleveland.Cholesterol[cleveland.Heart_disease=='No heart Disease'],  c="lightskyblue") 

plt.title("Heart Disease w.r.t Age and Serum Cholestoral")
plt.xlabel("Age")
plt.legend(["Heart Disease", "No heart Disease"])
plt.ylabel("Serum Cholestoral")
```
